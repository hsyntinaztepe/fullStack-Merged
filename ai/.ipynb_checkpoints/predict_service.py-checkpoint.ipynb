```python
from flask import Flask, request, jsonify
import joblib
import pandas as pd
import numpy as np
import traceback

# --- Model ve ön işleme nesnelerini yükle ---
model = joblib.load("uçak_davranış_modeli.pkl")
scaler = joblib.load("scaler.pkl")
label_encoders = joblib.load("label_encoders.pkl")
app = Flask(__name__)

# --- Ön işleme fonksiyonu ---
def preprocess(data):
    df = pd.DataFrame([data])
    
    # Kategorik sütunları encode et
    for col, le in label_encoders.items():
        # None veya NaN değerleri "UNKNOWN" yap
        df[col] = df[col].fillna("UNKNOWN").astype(str)
        
        # Eğitimde olmayan değer gelirse UNKNOWN'a çek
        df[col] = df[col].apply(lambda x: x if x in le.classes_ else "UNKNOWN")
        
        # UNKNOWN sınıfını ekle (eğer yoksa)
        if "UNKNOWN" not in le.classes_:
            # NumPy array olarak güncelle (liste değil!)
            new_classes = np.append(le.classes_, "UNKNOWN")
            le.classes_ = new_classes
        
        df[col] = le.transform(df[col])
    
    # Özellikleri seç (eğitimde kullandığın X ile aynı)
    X = df.drop(['id1', 'id2', 'suspicious'], axis=1, errors='ignore')
    
    # Ölçekleme
    X_scaled = scaler.transform(X)
    return X_scaled

# --- API endpoint ---
@app.route("/predict", methods=["POST"])
def predict():
    try:
        data = request.json
        print("\n[DEBUG] Gelen veri:", data)
        
        X_scaled = preprocess(data)
        pred = int(model.predict(X_scaled)[0])
        proba = float(model.predict_proba(X_scaled)[0][1])
        
        return jsonify({
            "prediction": pred,
            "probability": proba
        })
    except Exception as e:
        print("\n[ERROR] Tahmin sırasında hata oluştu!")
        print("[ERROR] Gelen veri:", request.json)
        traceback.print_exc()
        return jsonify({"error": str(e)}), 400

if __name__ == "__main__":
    app.run(host="0.0.0.0", port=5000)
```
